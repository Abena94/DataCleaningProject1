```python

# Remove newline characters 

df['Hits'] = df['Hits'].str.replace('\n', ' ')
df['Hits']=df['Hits'].str.replace('[^0-9]', '', regex=True)
df['Hits'] = df['Hits'].astype('int64')
df.head()
```

```python
# Split the 'Team Contract' column into 'Team' and 'Contract' columns
# creating two new columns and initially assigning them the same values as the Team & Contract column
df['Club'] = df['Team & Contract']  
df['Contract'] = df['Team & Contract']
def changes(row):
    a = row['Team & Contract'].split('\n') # splitting the team and contract column into a list based on the newline character
    ele=''
    x=[i for i in a if i!=ele]
    row['Club'] = x[0] # assigning the first element of the list to the club column
    row['Contract'] = x[1]  # assigning the second element to the contract column

    
    return row

df = df.apply(changes,axis='columns')
#f['Team'] = df['Team'].str.replace('\n', ' ')
#f['Contract'] = df['Contract'].str.replace('\n', ' ')
df.drop(['Team & Contract'], axis='columns',inplace=True) # The Team & Contract column is no longer required so we just go ahead and drop it.
    
df.head()
```

```python
#grouping each unique values in contract to kow their value count of each.
def values_contract(value):
    if 'On Loan' in value:
        return 'On Loan'
    elif 'Free' in value:
        return 'Free Players'
    elif '~' in value:
        return 'Under Contract'
    else:
        return 'Not Found'
```

```python
# this creates a new column using our check_contract function
df['contract_type'] = df['Contract'].apply(values_contract)
```

```python
# Year at which contract end

# The last four digit indicated the year the contract end
# while for the values similar to this 'Jan 1, 2021 On Loan', will be using slicing to obtain the year

def values(value):
    if 'Loan' in value:
        value = value[7:12]
        value = value.strip(' ') # I added this line because some whitespace can present
        return value
    else:
        value = value[-4:]
        return value
        
df['Contract_end'] = df['Contract'].apply(lambda x : values(x))
```

```python
#importing libraries that helps to check special characters in a column
import string
import re
alphabet = string.ascii_letters+string.punctuation
```

```python
#To check if there are special characters in Name and LongName column of the players
df.LongName.str.strip(alphabet).astype(bool).any()
```

```python
#Removing the special character in LongName using a function.
def strip_character(dataCol):
    r = re.compile(r'[^a-zA-Z !@#$%&*_+-=|\:";<>,./()[\]{}\']')
    return r.sub('', dataCol)

df['LongName'] = df['LongName'].apply(strip_character)
```

```python
# Define functions to extract numerical values from strings

# Define a function to convert heights in feet and inches to centimeters
def convert_height(height_string):
    try:
        # Split the height string into feet and inches
        feet, inches = map(int, height_string.split("'"))
        
        # Convert feet and inches to centimeters (1 foot = 30.48 cm, 1 inch = 2.54 cm)
        height_cm = (feet * 30.48) + (inches * 2.54)
        
        return height_cm
    except ValueError:
        return None

def extract_height(height_str):
    try:
        
        # Remove ' " ' from the string and call the function that convert it to an integer in cm 
        return int(convert_height((height_str.strip('"'))))
    except ValueError:
        # If there's an error (e.g., non-numeric value), return None
        return None

def extract_weight(weight_str):
    try:
        # Remove 'lbs' from the string and convert to an integer
        return int(weight_str.strip('lbs'))
    except ValueError:
        # If there's an error (e.g., non-numeric value), return None
        return None
```

```python
# Apply the functions to create new numerical columns
# We'll apply the extract_height function to the 'Height' column and create a new 'Height (cm)' column.
df['Height (cm)'] = df['Height'].apply(extract_height)

# Similarly, we'll apply the extract_weight function to the 'Weight' column and create a new 'Weight (kg)' column.
df['Weight (lbs)'] = df['Weight'].apply(extract_weight)
```

```python
#Drop the original non-numerical columns 
df = df.drop(['Height', 'Weight'], axis=1)
```

```python
# Convert the "Joined" column to datetime
df['Joined'] = pd.to_datetime(df['Joined'], format='%b %d, %Y')
```

```python
# Calculate the current date
current_date = datetime.now()
```

```python
# Calculate the time difference between the current date and "Joined"
df['Years Since Joining the current club'] = (current_date - df['Joined']).dt.days / 365
```

```python
# Check if someone enrolled more than 10 years ago
df['Joined More Than 10 Years Ago'] = df['Years Since Joining the current club'] > 10

# Display the DataFrame
df.head()
```

```python
df['Value_Euro'] = df['Value'].str.strip('€')
df['Release Clause_Euro'] = df['Release Clause'].str.strip('€')
df['Wage_Euro'] = df['Wage'].str.strip('€')
```

```python
def convert_to_dollar(dollar):
    if 'M' in dollar:
        dollar = dollar.strip('M')
        millon = float(dollar) * 1000000
        return millon
    else:
        dollar = dollar.strip('K')
        thousand = float(dollar) * 1000
        return thousand
```

```python
df['Value_Euro'] = df['Value_Euro'].apply(lambda X: convert_to_dollar(X))
df['Release Clause_Euro'] = df['Release Clause_Euro'].apply(lambda X: convert_to_dollar(X))
df['Wage_Euro'] = df['Wage_Euro'].apply(lambda X: convert_to_dollar(X))
```

```python
df.drop(['Value', 'Release Clause', 'Wage'], axis=1, inplace = True )
```

```python
def extract_star(star_str):
    try:
        # Remove '*' from the string and convert to an integer
        return int(star_str.strip('★'))
    except ValueError:
        # If there's an error (e.g., non-numeric value), return None
        return None
```

```python
df['IR'] = df['IR'].apply(extract_star)
df['WF']= df['WF'].apply(extract_star)
df['SM'] = df['SM'].apply(extract_star)
```
